```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline
```

```python
## now we open our dataset
```

```python
dataset = pd.read_excel("nhub_datasets.xlsx")
```

```python
## Cleaning the data, 
## so we replace the M and F in the gender column with 'Male' or 'Female'
```

```python
dataset["Gender"] = dataset["Gender"].str.replace('M', 'Male').str.replace('F', 'Female')
```

```python
# Combine all 'Tech Stack Enrolled' values into a single string
all_tech_stacks = ' '.join(dataset['Tech Stack Enrolled'].astype(str))

# Create a WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_tech_stacks)

# Display the WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Tech Stacks Enrolled')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(range(len(group_count3)), group_count3['Count'], tick_label=group_count3['Tech Stack Enrolled'] + ' - ' + group_count3['Program'] + ' - ' + group_count3['Gender'])
plt.xlabel('Tech Stack Enrolled - Program - Gender')
plt.ylabel('Count')
plt.title('Count of Combinations')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
grouped_count2 = dataset.groupby(['Tech Stack Enrolled', 'Program']).size().reset_index(name='Count')
tech_stacks = grouped_count2['Tech Stack Enrolled'].unique()
plt.figure(figsize=(16, 8))

for idx, tech_stack in enumerate(tech_stacks, start=1):
    plt.subplot(1, len(tech_stacks), idx)
    
    # Filter the data for the current 'Tech Stack Enrolled' category
    subset = grouped_count2[grouped_count2['Tech Stack Enrolled'] == tech_stack]
    
    # Create a bar chart
    plt.bar(subset['Program'], subset['Count'], tick_label=subset['Program'])
    plt.xlabel('Program')
    plt.ylabel('Count')
    plt.title(tech_stack, fontsize=12)  # Adjust the font size for the title

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust the top parameter for space
plt.suptitle('Program Distribution by Tech Stack Enrolled', fontsize=14)
plt.show()
```

```python
# Grouping by 'Tech Stack Enrolled' and 'Program', and getting the count
grouped_count2 = dataset.groupby(['Tech Stack Enrolled', 'Program']).size().reset_index(name='Count')

# Create a separate pie chart for each 'Tech Stack Enrolled' category
tech_stacks = grouped_count2['Tech Stack Enrolled'].unique()
plt.figure(figsize=(40, 20))

for tech_stack in tech_stacks:
    subset = grouped_count2[grouped_count2['Tech Stack Enrolled'] == tech_stack]
    plt.subplot(1, len(tech_stacks), tech_stacks.tolist().index(tech_stack) + 1)
    plt.pie(subset['Count'], labels=subset['Program'], autopct='%2.2f%%', startangle=140, textprops={'fontsize': 20})
    plt.title(tech_stack, fontsize=20)

plt.tight_layout(rect=[0, 0.03, 1, 0.85])
plt.suptitle('Program Distribution by Tech Stack Enrolled', fontsize=30)
plt.legend()
plt.show()
```

```python
group_count3 = dataset.groupby(['Tech Stack Enrolled','Gender']).size().reset_index(name='Count')

# Create subplots for each 'Tech Stack Enrolled' category
tech_stacks = group_count3['Tech Stack Enrolled'].unique()
plt.figure(figsize=(16, 8))

colors = {'Male': 'olive', 'Female': 'purple'}

for idx, tech_stack in enumerate(tech_stacks, start=1):
    plt.subplot(1, len(tech_stacks), idx)
    
    # Filter the data for the current 'Tech Stack Enrolled' category
    subset = group_count3[group_count3['Tech Stack Enrolled'] == tech_stack]
    
    # Create a bar chart
    plt.bar(subset['Gender'], subset['Count'], color=[colors[g] for g in subset['Gender']])
    plt.xlabel('Gender')
    plt.ylabel('Count')
    plt.title(tech_stack, fontsize=12)  # Adjust the font size for the title

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.suptitle('Program and Gender Distribution by Tech Stack Enrolled', fontsize=14)
plt.show()
```
